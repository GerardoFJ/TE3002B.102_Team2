## 4. ML #1 — K-Means: orden óptimo de transferencia del xArm

**Problema**: el xArm 6 está fijo en `(11.5, 3.35)` y debe transferir
los 3 PuzzleBots desde el dorso del ANYmal (3 puntos) a 3 destinos en
la mesa (3 puntos). El orden hardcoded actual es `C → B → A`. Pero,
¿es el orden óptimo en términos de **distancia recorrida por el
efector**?

**Idea**: usamos K-Means con `k = 3` sobre los 6 puntos (3 picks + 3
drops). Cada cluster contiene 1 pick + 1 drop "naturalmente" cercanos
(la geometría del problema lo favorece). El orden de visita de los
clusters se decide por **distancia mínima desde el home del xArm
hacia el centroide del cluster**.

Esto da una asignación pick↔drop y un orden de visita en una sola
pasada — es el corazón del concepto **multi-robot task allocation**
basado en clustering.

````python
from sklearn.cluster import KMeans

# 1. Posiciones de PBs sobre el ANYmal (final fase 2)
ax_, ay_, ayaw = (p2['base_x'][-1], p2['base_y'][-1], p2['base_yaw'][-1])
pb_picks = {}
for role in ('A', 'B', 'C'):
    off = coord.PB_ON_ANYMAL_OFFSETS[role]
    cs, sn = math.cos(ayaw), math.sin(ayaw)
    pb_picks[role] = np.array([
        ax_ + cs*off[0] - sn*off[1],
        ay_ + sn*off[0] + cs*off[1],
        off[2],
    ])

# 2. Drops en la mesa
pb_drops = {role: np.array([d[0], d[1], 0.10])
            for role, d in coord.PB_TABLE_DROP.items()}

# 3. Stack de los 6 puntos para clustering (etiquetamos cada uno)
points = []
labels_text = []
for role, p in pb_picks.items():
    points.append(p);  labels_text.append(f'pick_{role}')
for role, p in pb_drops.items():
    points.append(p);  labels_text.append(f'drop_{role}')
X = np.array(points)

# 4. K-Means con 3 clusters
km = KMeans(n_clusters=3, n_init=10, random_state=42).fit(X)
cluster_id = km.labels_
centroids = km.cluster_centers_

# 5. Para cada cluster, ¿qué role contiene? (debería ser el mismo en pick+drop)
print('Asignación K-Means (cluster -> puntos):')
for c in range(3):
    members = [labels_text[i] for i in range(len(labels_text))
               if cluster_id[i] == c]
    centroid_str = f'({centroids[c,0]:.2f}, {centroids[c,1]:.2f}, {centroids[c,2]:.2f})'
    print(f'  cluster {c}: {members} -> centroid {centroid_str}')

# 6. Orden de visita: distancia desde home del xArm a cada centroide
home_w = coord.xarm.world_from_arm(
    coord.xarm.forward_kinematics(coord.xarm.q_home))
dists = np.linalg.norm(centroids - home_w, axis=1)
visit_order = np.argsort(dists)
print(f'\nDistancias home→centroide: {dists.round(3)}')
print(f'Orden óptimo de visita (de cluster más cercano al más lejano): {visit_order}')

# 7. Traducir el orden de visita a roles (cada cluster contiene un role)
ordered_roles = []
for c in visit_order:
    members = [labels_text[i] for i in range(len(labels_text))
               if cluster_id[i] == c]
    role = members[0].split('_')[1]   # 'pick_X' o 'drop_X' -> 'X'
    ordered_roles.append(role)
print(f'Orden óptimo C/B/A: {ordered_roles}')
print(f'Orden hardcoded:    {list(coord.TRANSFER_ORDER)}')
````

````python
# Visualización del clustering en el plano (x, y)
fig, ax = plt.subplots(figsize=(9, 5))
colors = ['#e41a1c', '#377eb8', '#4daf4a']
markers_pick = 'o'
markers_drop = 's'

for i, lbl in enumerate(labels_text):
    c = cluster_id[i]
    marker = markers_pick if lbl.startswith('pick') else markers_drop
    ax.scatter(X[i, 0], X[i, 1], c=colors[c], marker=marker, s=140,
               edgecolor='black', linewidth=1.2, zorder=3)
    ax.text(X[i, 0]+0.03, X[i, 1]+0.03, lbl, fontsize=8)

ax.scatter(centroids[:, 0], centroids[:, 1], c=colors, marker='*',
           s=300, edgecolor='black', linewidth=1.5, zorder=4,
           label='centroide')

# Marca xArm home
ax.scatter(home_w[0], home_w[1], c='gold', marker='X', s=200,
           edgecolor='black', linewidth=1.2, label='xArm home')

# Flechas del orden de visita
for k in range(len(visit_order) - 1):
    a = centroids[visit_order[k]][:2]
    b = centroids[visit_order[k+1]][:2]
    ax.annotate('', xy=b, xytext=a,
                arrowprops=dict(arrowstyle='->', color='black', lw=1.5))

ax.set_xlim(10.5, 12.0); ax.set_ylim(2.7, 4.0)
ax.set_aspect('equal')
ax.set_xlabel('x [m]'); ax.set_ylabel('y [m]')
ax.set_title('K-Means (k=3) sobre picks + drops + orden por distancia desde xArm home')
ax.legend(loc='upper left')
plt.tight_layout()
plt.show()
````

## 5. ML #2 — Ridge Regression: tiempo de despeje vs velocidad máxima

**Problema**: ¿cuánto tiempo tarda el Husky en despejar el corredor
en función de su velocidad máxima `v_max`? Queremos un modelo
predictivo que sirva para **planificar la misión completa**: si
acelero el robot, ¿cuánto baja `T_clear`?

Físicamente esperamos $T_{\text{clear}} \approx a + b/v_{\max}$
(tiempo inversamente proporcional a la velocidad). Probamos esto con
**Ridge sobre features polinomiales**: dejamos que el modelo
descubra la relación.

**Setup**: corremos N=20 simulaciones variando `v_max` en [0.25, 0.85] m/s.
Features: `[v_max, 1/v_max]`. Modelo: Ridge con $\lambda = 0.1$.

````python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_squared_error

# Generamos N corridas variando la velocidad maxima del controlador
N_RUNS = 20
vmaxs = np.linspace(0.25, 0.85, N_RUNS)

t_clear = np.zeros(N_RUNS)
for k, vm in enumerate(vmaxs):
    husky = HuskyA200(); husky.set_terrain('grass')
    husky.reset(x=0.0, y=2.0, theta=0.0)
    world = CorridorWorld()
    pusher = HuskyPusher(husky, world)
    pusher.controller.v_max = float(vm)
    log_t = pusher.clear_corridor(dt=0.05)
    t_clear[k] = log_t['t'][-1]

# Features: el modelo tiene que descubrir que T ~ 1/v_max
X = np.column_stack([vmaxs, 1.0 / vmaxs])
y = t_clear

# Split
np.random.seed(0)
idx = np.random.permutation(N_RUNS)
split = int(0.8 * N_RUNS)
i_tr, i_te = idx[:split], idx[split:]

ridge = Ridge(alpha=0.1).fit(X[i_tr], y[i_tr])
yhat_tr = ridge.predict(X[i_tr])
yhat_te = ridge.predict(X[i_te])

print(f'Ridge (alpha=0.1)')
print(f'  Train R^2: {r2_score(y[i_tr], yhat_tr):+.3f}')
print(f'  Test  R^2: {r2_score(y[i_te], yhat_te):+.3f}')
print(f'  Test RMSE: {mean_squared_error(y[i_te], yhat_te)**0.5:.3f} s')
print(f'\nCoeficientes:')
print(f'  beta_v_max (lineal):  {ridge.coef_[0]:+.3f}')
print(f'  beta_1/v_max (inv):   {ridge.coef_[1]:+.3f}  <-- el termino dominante')
print(f'  intercept:            {ridge.intercept_:+.3f}')
````

````python
# Plot: scatter + curva ajustada en el espacio v_max -> T_clear
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

# Panel izquierdo: T_clear vs v_max
ax = axs[0]
v_grid = np.linspace(0.20, 0.90, 100)
X_grid = np.column_stack([v_grid, 1.0 / v_grid])
y_grid = ridge.predict(X_grid)
ax.plot(v_grid, y_grid, '-', color='#377eb8', linewidth=2,
        label='Ridge: a + b·v + c/v')
ax.scatter(vmaxs[i_tr], y[i_tr], c='#4daf4a', s=70, edgecolor='black',
           label='train', zorder=3)
ax.scatter(vmaxs[i_te], y[i_te], c='#e41a1c', s=70, marker='^',
           edgecolor='black', label='test', zorder=3)
ax.set_xlabel('v_max [m/s]')
ax.set_ylabel('tiempo de despeje T_clear [s]')
ax.set_title('Husky: v_max -> T_clear (relacion ~ 1/v_max)')
ax.legend(loc='upper right')

# Panel derecho: predicho vs real (recta y=x)
ax = axs[1]
yhat_all = ridge.predict(X)
ax.scatter(y[i_tr], yhat_all[i_tr], c='#4daf4a', s=70, edgecolor='black',
           label='train', zorder=3)
ax.scatter(y[i_te], yhat_all[i_te], c='#e41a1c', s=70, marker='^',
           edgecolor='black', label='test', zorder=3)
lim = [y.min() * 0.95, y.max() * 1.05]
ax.plot(lim, lim, 'k--', alpha=0.6, label='y = y_hat')
ax.set_xlabel('T_clear real [s]')
ax.set_ylabel('T_clear predicho [s]')
ax.set_title('Calidad del fit: predicho vs real')
ax.legend(loc='upper left')
plt.tight_layout()
plt.show()
````

## 6. ML #3 — Random Forest: detector de singularidades del ANYmal

**Problema**: el modelo simplificado de las patas del ANYmal tiene
$\det(J) \propto \sin(q_1)\sin(q_3)$. Si $q_1 \to 0$ o $q_3 \to 0$ la
pata es **singular** y el control de fuerza diverge. Antes de
arrancar la fase 2 nos gustaría predecir si una configuración inicial
de gait va a meternos en problemas.

**Dataset sintético**: muestreamos 2000 configuraciones aleatorias
de `home_q = (q1, q2, q3)` dentro de rangos realistas y simulamos un
paso del trote, midiendo si alguna pata acaba con `|det(J)| < 1e-3`.

**Random Forest (100 árboles)** entrena binario:
$y = 1$ si la config es singular, $0$ si está limpia.

Bonus: el RF nos da la **importancia de cada articulación** —
veremos si confirma que `q1` y `q3` son los que más pesan.

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Generar dataset sintético: aleatorio (q1,q2,q3), corremos 1 ciclo de
# gait corto, etiquetamos como "singular" si min |det(J)| < 1e-3
def sample_config():
    return np.array([
        np.random.uniform(-0.6, 0.6),    # q1 (incluye cerca de 0 -> singular)
        np.random.uniform(0.3, 1.0),     # q2
        np.random.uniform(-1.7, -0.05),  # q3 (incluye cerca de 0 -> singular)
    ])

N = 800
X_rf = []
y_rf = []
for _ in range(N):
    q = sample_config()
    bot = ANYmal()
    try:
        gait = ANYmalTrotGait(bot, home_q=q)
        # Simular 0.5s de gait y registrar min|det J|
        min_det = np.inf
        for t_ in np.arange(0, 0.5, 0.01):
            _, det_jacs, _ = gait.step(t_, v_forward=0.2)
            for det in det_jacs.values():
                if abs(det) < min_det:
                    min_det = abs(det)
        is_singular = int(min_det < 1e-3)
    except Exception:
        is_singular = 1   # config no válida
    X_rf.append(q)
    y_rf.append(is_singular)

X_rf = np.array(X_rf);  y_rf = np.array(y_rf)
print(f'Dataset RF: {len(X_rf)} muestras, '
      f'{int(y_rf.sum())} singulares ({100*y_rf.mean():.1f}%)')

X_tr, X_te, y_tr, y_te = train_test_split(
    X_rf, y_rf, test_size=0.2, random_state=42, stratify=y_rf)

rf = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_tr, y_tr)
y_pred = rf.predict(X_te)
print(f'Accuracy test: {accuracy_score(y_te, y_pred):.3f}')
print('\nClassification report:')
print(classification_report(y_te, y_pred, target_names=['safe', 'singular']))
print('\nFeature importance:')
for f, imp in zip(['q1', 'q2', 'q3'], rf.feature_importances_):
    bar = '█' * int(imp * 40)
    print(f'  {f}: {imp:.3f}  {bar}')
````

````python
# Visualización: scatter q1 vs q3 (los 2 que importan según teoría),
# coloreado por etiqueta real, contornos de probabilidad del RF.
fig, axs = plt.subplots(1, 2, figsize=(13, 5))

# Panel izquierdo: dataset etiquetado (q1 vs q3)
ax = axs[0]
mask_safe = (y_rf == 0)
mask_sing = (y_rf == 1)
ax.scatter(X_rf[mask_safe, 0], X_rf[mask_safe, 2], c='#4daf4a', s=10,
           alpha=0.6, label='safe')
ax.scatter(X_rf[mask_sing, 0], X_rf[mask_sing, 2], c='#e41a1c', s=10,
           alpha=0.6, label='singular')
ax.axvline(0, color='black', linestyle='--', alpha=0.5, linewidth=1)
ax.axhline(0, color='black', linestyle='--', alpha=0.5, linewidth=1)
ax.set_xlabel('q1 [rad]'); ax.set_ylabel('q3 [rad]')
ax.set_title('Dataset: q1 vs q3 (etiqueta real)')
ax.legend()

# Panel derecho: confusion matrix
ax = axs[1]
cm = confusion_matrix(y_te, y_pred)
im = ax.imshow(cm, cmap='Blues')
for i in range(2):
    for j in range(2):
        ax.text(j, i, cm[i, j], ha='center', va='center',
                color='white' if cm[i, j] > cm.max()/2 else 'black',
                fontsize=14, fontweight='bold')
ax.set_xticks([0, 1]); ax.set_xticklabels(['safe', 'singular'])
ax.set_yticks([0, 1]); ax.set_yticklabels(['safe', 'singular'])
ax.set_xlabel('predicho'); ax.set_ylabel('real')
ax.set_title('Confusion Matrix (test set)')
plt.colorbar(im, ax=ax)
plt.tight_layout()
plt.show()
````

## 7. Coordinación a tres niveles

El sistema combina ML local en cada robot con una FSM central. Esto
nos da **3 niveles de coordinación**:

### Nivel 1 — Local (dentro de cada robot)
Cada robot tiene su propia lógica de control:

- **Husky**: `HuskyPusher` decide qué caja empujar y la trayectoria
  con LiDAR 2D simulado.
- **ANYmal**: `ANYmalTrotGait` genera referencias del pie por pata e
  IK 2-link planar; `ANYmalNavigator` produce `(v, ω)` para la base.
- **xArm 6**: trayectoria cartesiana con waypoints + IK 6-DoF (4 DoF
  efectivos en el modelo simplificado).

### Nivel 2 — Handoff (FSM `MissionCoordinator`)
La máquina de estados secuencia las fases y pasa el contexto:

```
HUSKY_CLEAR ─pose final + cajas─▶ ANYMAL_TRANSPORT ─pose final ANYmal─▶
XARM_TRANSFER ─PBs en mesa─▶ DONE
```

El coordinator es **dumb**: solo invoca métodos de los módulos
especializados (siguiendo el consejo del profe).

### Nivel 3 — Meta (ML asistiendo decisiones)
ML inyecta inteligencia en puntos clave:

| Decisión               | Método ML       | Lo que reemplaza             |
|------------------------|-----------------|------------------------------|
| Orden xArm pick-place  | K-Means (k=3)   | Hardcoded `C → B → A`        |
| Predicción `v_cmd`     | Ridge           | Heurística proporcional      |
| Pre-flight singularity | Random Forest   | Asunción ciega (`q1=0.25`)   |

Los 3 modelos se entrenan con **datos sintéticos generados por la
propia simulación** — un patrón típico de Sim2Real.

````python
# Diagrama del flujo de coordinación (matplotlib puro, sin graphviz)
fig, ax = plt.subplots(figsize=(13, 6))
ax.set_xlim(0, 10); ax.set_ylim(0, 6)
ax.axis('off')

def box(ax, x, y, w, h, text, color='#cce5ff', edge='black'):
    rect = mpatches.FancyBboxPatch((x, y), w, h, boxstyle='round,pad=0.05',
                                    facecolor=color, edgecolor=edge,
                                    linewidth=1.5)
    ax.add_patch(rect)
    ax.text(x + w/2, y + h/2, text, ha='center', va='center', fontsize=10,
            fontweight='bold')

def arrow(ax, x1, y1, x2, y2, label=''):
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(arrowstyle='->', color='black', lw=1.5))
    if label:
        ax.text((x1+x2)/2, (y1+y2)/2 + 0.1, label, ha='center', fontsize=8,
                style='italic')

# Nivel 1
ax.text(0.2, 5.4, 'NIVEL 1 — Local', fontsize=11, fontweight='bold', color='#0033aa')
box(ax, 0.5, 4.3, 1.8, 0.8, 'Husky\nPusher',  '#cce5ff')
box(ax, 2.7, 4.3, 1.8, 0.8, 'ANYmal\nGait',   '#e3d4f1')
box(ax, 4.9, 4.3, 1.8, 0.8, 'xArm 6\nIK',     '#dddddd')

# Nivel 2
ax.text(0.2, 3.4, 'NIVEL 2 — Handoff', fontsize=11, fontweight='bold', color='#aa3300')
box(ax, 1.5, 2.4, 5.0, 0.8, 'MissionCoordinator\n(FSM: HUSKY_CLEAR → ANYMAL_TRANSPORT → XARM_TRANSFER → DONE)',
    '#ffe0b3')
arrow(ax, 1.4, 4.3, 2.5, 3.2)
arrow(ax, 3.6, 4.3, 4.0, 3.2)
arrow(ax, 5.8, 4.3, 5.5, 3.2)

# Nivel 3
ax.text(0.2, 1.5, 'NIVEL 3 — Meta (ML)', fontsize=11, fontweight='bold', color='#118811')
box(ax, 0.5, 0.4, 1.8, 0.8, 'K-Means\norden pick-place',  '#bce4b3')
box(ax, 2.7, 0.4, 1.8, 0.8, 'Ridge\nv_cmd modelo',         '#bce4b3')
box(ax, 4.9, 0.4, 1.8, 0.8, 'Random Forest\nsingularity',  '#bce4b3')

# Flechas: ML -> Coordinator
arrow(ax, 1.4, 1.2, 3.0, 2.4, '')
arrow(ax, 3.6, 1.2, 4.0, 2.4, '')
arrow(ax, 5.8, 1.2, 5.0, 2.4, '')

# Caja final
box(ax, 7.2, 2.8, 2.5, 0.8, 'misión\ncompletada ✓', '#a4d3a4')
arrow(ax, 6.5, 2.8, 7.2, 3.2)

ax.set_title('Coordinación a 3 niveles del sistema multi-robot',
             fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
````
